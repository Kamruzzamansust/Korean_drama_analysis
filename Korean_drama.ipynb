```python
def has_desired_day(days_list, desired_day):
    if pd.isna(days_list):
        return False
    days = days_list.split(', ')
    return desired_day in days

# Define the days of the week
days_of_week = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

# Iterate through years and days
for year in range(2015, 2024):
    print(f"Year: {year}")
    for day in days_of_week:
        filtered_df = df[(df['aired_on'].apply(has_desired_day, args=(day,))) & (df['year'] == year)]
        num_entries = filtered_df.shape[0]
        print(f"{day}: {num_entries} entries")
```
